```python
import numpy
xn = numpy.array([[1, 1, 1], [0, 0, 0], [1, 1, 0], [1, 0, 0], [0, 1, 0],[1, 0, 1]])
print(xn)
dim = 3
PopSize = 6
diversity_list = []
centroid = numpy.zeros(dim)
I = 0
for i in range (0,dim):
    for j in range(0,PopSize):
        centroid[i] += xn[j,i] / PopSize

print("centroid", centroid)           
for i in range (0, dim):
    for j in range (0, PopSize):
        I += (xn[j,i] - centroid[i])**2
diversity_list.append(float(I))
print("diversity = ", diversity_list)
```
